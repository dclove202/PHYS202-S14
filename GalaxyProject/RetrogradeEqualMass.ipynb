# Retrograde Galaxy Interaction With Equal Masses

```python
%pylab inline
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
```

### Placing The Stars in Orbit

```python
## Creating Arrays for Star Postitions
def ring(ringnumber):
    '''
    Since their are multiple rings in the star field surrounding the galaxy,
    I thought I would make a function that outputs the distance, number of stars,
    x positions, y positions, and the angle from the origin for each star in a ring.
    '''
    r = ((ringnumber + 1.)/10.)*25.
    nstars = 6.*(ringnumber+1.)
    theta = (2*np.pi)/nstars
    xpositions = np.zeros(nstars)
    ypositions = np.zeros(nstars)
    for i in range(6*(ringnumber+1)):
        xpositions[i] = r*np.cos(i*theta) 
        ypositions[i] = r*np.sin(i*theta)
    return float(r),int(nstars),xpositions, ypositions,theta

#Initializing Lists
xlst = []
ylst = []
xvel = []
yvel = []
thet = []

#Iterating over all 5 rings in the star system
for i in range(1,6):
    r,n,xx,yy,theta = ring(i)
    G = 4.499736* 10**-8
    M = 10.**11
    r = r
    #Appending the initialized lists with the values from the function,
    #as well as calculating the velocities for the particles.
    
    for b in range(n):
        thet.append(theta*b)
        xvel.append(np.sqrt(G*M/r)*np.cos((theta*b)+(np.pi/2)))
        yvel.append(np.sqrt(G*M/r)*np.sin((theta*b)+(np.pi/2)))
        xlst.append(xx[b])
        ylst.append(yy[b])
        
        #Plotting all of the stars to make sure they look as they should
        plt.scatter(xx[b],yy[b])


th = np.array(thet)
xp = np.array(xlst)
yp = np.array(ylst)
XV = np.array(xvel)
YV = np.array(yvel)
plt.axis('equal')
plt.show()
        


```

### odeint Description of Masless Stellar Orbits

```python
'''
In this cell, I create an initial position array and a function and pass them
into odeint.
'''

## Defining Constants
M = 10.0**11.00
G = 4.5* 10**-8

#Initializing Lists
xvp = []
yvp = []
a = np.pi/2 # A constant I will use in calculating velocities

#Calculating tangential velocities and adding them to arrays by component
for i in range(120):
    vy = np.sqrt(G*M/(np.sqrt(xp[i]**2+yp[i]**2)))*np.cos(th[i]+a)
    yvp.append(vy)
for i in range(120):
    vx = np.sqrt(G*M/(np.sqrt(xp[i]**2+yp[i]**2)))*np.sin(th[i]+a)
    xvp.append(vx)
yvp = np.array(yvp)
xvp = np.array(yvp)


#Creating an initial array for odeint of form[x,y,vx,vy]
#Iterating through every star and velocity
for i in range(120):
    initials = [xp[i],yp[i],XV[i],YV[i]]
    time = np.arange(0.,1.,1.)
#Function to feed odeint
    def motion(initials,time):
        '''
        A function that takes in initial conditions and 
        a time array, and returns an array with the 
        derivatives of those initial conditions.
        '''
        a = np.zeros(4)
        a[0] = initials[2]
        a[1] = initials[3]
        R = np.sqrt(initials[0]**2 + initials[1]**2)
        a[2] = (-G*M*initials[0]/R**3)
        a[3] = (-G*M*initials[1]/R**3)
        return a
#Odeint
    w = odeint(motion,initials,time)
    xp1 = w[:,0]
    yp1 = w[:,1]
    xvp1 = w[:,2]
    yvp1 = w[:,3]
    
    #Plotting the x and y values with arrows pointing
    #in the direction of their velocities.
    plt.quiver(xp1,yp1,xvp1,yvp1)
plt.axis('equal')
plt.show()
```

```python
def disruption(moment,t):
    '''
    This function returns the derivatives for the 8 part initial conditions
    array that consists of the positions and velocities of the massless stars.
    The function depends on both the intial conditions involved and the 
    differential equations from pfleiderer's paper.
    '''
    g = np.zeros(8)
    g[0] = moment[2]
    g[1] = moment[3]
    #r
    rxx = moment[0]
    ryy = moment[1]
    r = np.sqrt(rxx**2 + ryy**2)
    #R
    Ry = moment[5]
    Rx = moment[4]
    R = np.sqrt(Rx**2 + Ry**2)
    #Rho
    rhox = Rx - rxx
    rhoy = Ry - ryy
    rho = np.sqrt(rhox**2+rhoy**2)
    
    g[2] = -G*((M*rxx/(r**3))-(S*rhox/(rho**3))+(S*Rx/(R**3)))
    g[3] = -G*((M*ryy/(r**3))-(S*rhoy/(rho**3))+(S*Ry/(R**3)))
    g[4] = moment[6]
    g[5] = moment[7]
    g[6] = -G*(M+S)*Rx/R**3
    g[7] = -G*(M+S)*Ry/R**3
    return g
```

```python
def Smash(xp,yp,XV,YV,x,y,vx,vy):
    '''
    This function was made so that I could iterate through the intial conditions
    given from other cells and pass them into odeint. The function takes in 
    intial conditions and calls on the 'disruption' function to return arrays
    with the states of the galaxies and stars over the time interval specified.
    It's called smash because it smashes galaxies together.
    '''
    #Constants
    M = 10.0**11.00
    S = M
    G = 4.499736*10**-8
    #Time Array
    t = np.arange(0,11,1.)
    #Initial Condtions
    moment = [xp,yp,XV,YV,x,y,vx,vy]
    #Defining a variable as odeint
    rr = odeint(disruption,moment,t)
    #Defining variables as 2 dimensional time arrays that include
    #positions and velocities to return
    ppx = rr[:,0]
    ppy = rr[:,1]
    ppvx = rr[:,2]
    ppvy = rr[:,3]
    Sx = rr[:,4]
    Sy = rr[:,5]
    Svx = rr[:,6]
    Svy = rr[:,7]
    
    return ppx,ppy,ppvx,ppvy,Sx,Sy,Svx,Svy
```

### Test for Plotting The Galaxies (equal mass, parabolic passage)

```python
'''
This is my first attempt at plotting the interaction between galaxies.
For simplicity, I have not yet attached any of the rotating stars
to the 'victim' galaxy. In this cell, I define constants based on Toomres'
paper, and then use the differential equations from Pfleiderer's paper to
plot the motion using odeint.
'''
G = 4.499736* 10**-8 # G in kiloparsecs, solar masses, and millions of years.
                    # Intial conditions of disruptor mass
Rmin = 25.0         # Rmin, where we will start the particle (Xo)
M = 10**11         #Solar Masses
S = M
#Initial conditions for the disruptor galaxy
it = [-13.1401625858, 35.8051639777, 15.3609179799, -1.08701216176e-05]   

#Creating a time array
tinit = 0.
tfin = 10
timestep = .001
t = arange(tinit,tfin,timestep)   

def gmd(inits,time):
    '''
    This function takes inputs from the initial conditions of the disruptor galaxy
    as well as uses the ordinary differential equation from Pfleiderer's paper
    to put out an array with the deriviatives of the initial conditions. The
    purpose of this function is to pass it into odeint.
    'gmd' stands for 'Galaxy Motion Derivative'.
    '''
    f=zeros(4)
    f[0] = inits[2]
    f[1] = inits[3]
    R = sqrt(inits[0]**2 + inits[1]**2)
    f[2] = -(G*(M+S)*inits[0])/R**3
    f[3] = -(G*(M+S)*inits[1])/R**3
    return f

#Using odeint to return an array with [time,initial condition]
z = odeint(gmd,it,t) 

#Setting variables for the arrays of conditions
x = z[:,0]               
y = z[:,1]  
vx = z[:,2] 
vy = z[:,3]

#Plotting the motion of the disruptor galaxy
plt.scatter(0,0, s=100)
plt.scatter(Rmin,0, s=100)
plt.plot(x,y)
plt.axis('equal')
plt.show()

'''
These are good initial conditions. After some analysis, the galaxy starts
relatively far away and clearly moves to a minimum distance to the victim
of 25 kiloparsecs. Just as it happened according to Toomres' paper.
'''
```
