## Question 2

(a) Use Euler's method (not odeint) to solve the following first-order differential equation with given intitial condition y(0)=0.

dy/dx = x + 2y

Find the value of y when x=1, y(1), using two different step sizes: 0.25 then 0.02.

```python

def Euler_ODE(dx, sol):
    xmax = sol+2*dx
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    #Redundant Initial condition
    x[0]=0
    y[0]=0
    
    for i in range (nsteps-1):
        x[i+1] = x[i] + dx
        y[i+1]= y[i] + dx*(x[i] + 2*y[i])
    return x,y
x,y = Euler_ODE(.25,1)
print "For steps of .25, y(1) =", y[-1]
g,h = Euler_ODE(.02,1)
print "For steps of .02, y(1) =", h[-1]
```

(b) Plot the results from part (a) for the two different step sizes with the exact solution

y = 0.25e^2x -.5x -.25

on one plot. Include a legend. Which step size gives a more accurate result?

```python
t=np.arange(0,1+.01,.01)
func = lambda t: .25*e**(2*t) - .5*t - .25
plt.plot(t,func(t), label='Exact Solution',color='orange',linewidth=3)
plt.plot(x,y, label='.25 Step Size', color='green',linewidth=3)
plt.plot(g,h,label='.02 Step Size', color='black',linewidth=3)
plt.legend(fancybox=True, loc='best', framealpha=.4)
plt.show()
```

The smaller step size, .02, gives the more accurate result.

(c) Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def Euler_Cromer(dx, sol):
    xmax = sol+2*dx
    nsteps = int(xmax/dx)
    dydx = np.zeros(nsteps)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    
    for i in range (nsteps-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i]+2*y[i]
        y[i+1]= y[i] + dx*dydx[i+1]
    return x,y
```

```python
c,d = Euler_Cromer(.25,1)
k,l = Euler_Cromer(.02,1)
```

```python
def midpoint_method(dx, sol):
    xmax = sol+2*dx
    nsteps = int(xmax/dx)
    dydx = np.zeros(nsteps)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    for i in range(nsteps-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        y[i+1] = y[i] + .5*dx*(dydx[i]+dydx[i+1])
    return x, y
m,n = midpoint_method(.25,1)
o,p = midpoint_method(.02,1)
```

```python
t=np.arange(0,1+.01,.01)
func = lambda t: .25*e**(2*t) - .5*t - .25
plt.plot(t,func(t), label='Exact Solution',linewidth=3)
plt.plot(c,d, label='Euler-Cromer .25', linewidth=3)
plt.plot(k,l, label='Euler-Cromer .02', linewidth = 3)
plt.plot(m,n, label='Midpoint .25', linewidth = 3)
plt.plot(o,p, label='Midpoint .02', linewidth=3)

plt.legend(loc='best', fancybox=True, framealpha=.4)
plt.show()
```

```python
func = lambda t: .25*e**(2*t) - .5*t - .25
plt.plot(c,func(c)-d, label='Euler-Cromer .25', linewidth=3)
plt.plot(k,func(k)-l, label='Euler-Cromer .02', linewidth = 3)
plt.plot(m,func(m)-n, label='Midpoint .25', linewidth = 3)
plt.plot(o,func(o)-p, label='Midpoint .02', linewidth=3)
plt.legend(loc='best', fancybox=True, framealpha=.4)
plt.show()
```
