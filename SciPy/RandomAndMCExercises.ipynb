## Question 1

The Weibull distribution, with parameters α>0 and β>0 is described by the density function

f(x)=                  $f(x) = \alpha \beta^{-\alpha}x^{\alpha - 1} e^{-(x/\beta)^{\alpha}}$                          if x>0

    0                   otherwise

(a) Solve for the CDF of the Weibull distribution analytically, using
F(x)=∫x−∞f(t)dt

Using a U substitution: U = (1/α)*x^α, and integrating with x as the upper bound and 0 as the lower,
My solution is:

f(x) = $1 - e^{-(x/\beta)^{\alpha}}$

(b) Find the inverse of the CDF $F^{−1}(x)$ analytically.

It's inverse, found by substituting f^-1(x) for x and x for f(x) and solving, is:

$f^{-1}(x) = \beta(-ln(1-x))^{1/\alpha}$

(c) Write a program to generate 10,000 samples from the Weibull distribution and plot the true distribution, the CDF and the sampled distribution as a histogram, all on the same graph. Let 
α = 1.5, 
β = 6.
Note: Do not use the built-in Weibull function from numpy.random.

```python
#Functions
alpha = 1.5
beta = 6.0
x = np.arange(0,100,.01)
f = lambda x: alpha*beta**(-alpha) * x**(alpha-1) * np.e**(-(x/beta)**alpha)
F = lambda x: 1 - np.e**(-1*(x/beta)**alpha)
Finv = lambda z: beta*((-1*np.log(1-z))**(1./alpha))

#Plots
plt.plot(x,f(x),linewidth=2,label='True Distribution')
plt.plot(x,F(x),linewidth=2, label='Cumulative Function')
plt.xlim(0,20)
plt.ylim(0,1.1)

#Histogram
N = 10000
z = numpy.random.random_sample(N)
xsamp = Finv(z)
plt.hist(xsamp, 50, normed=True, label="Sampled Distribution")

plt.legend(loc='best',fancybox=True,framealpha=.3)
plt.xlabel(r"$X$", fontsize=20)
plt.ylabel(r'$f(x) = \alpha \beta^{-\alpha}x^{\alpha - 1} e^{-(x/\beta)^{\alpha}}$', fontsize = 15)
plt.show()

```
